```python
# Gerekli kütüphaneleri içe aktarıyoruz.
import turtle  # Grafik arayüzü ve çizimler için.
import time  # Oyunu yavaşlatmak ve zamanlama işlemleri için.
import random  # Yemi rastgele bir konuma yerleştirmek için.

# --- Oyunun Temel Ayarları ---
delay = 0.1  # Oyunun başlangıç hızı (küçük değer daha hızlı demek).
score = 0  # Oyuncunun mevcut skoru.
high_score = 0  # Ulaşılan en yüksek skor.
segment_size = 20  # Yılanın her bir parçasının ve hareket mesafesinin piksel boyutu.
screen_width = 600  # Oyun ekranının genişliği.
screen_height = 600  # Oyun ekranının yüksekliği.

# --- Ekran Kurulumu ---
screen = turtle.Screen()  # Bir ekran nesnesi oluşturuluyor.
screen.title("Yılan Oyunu")  # Pencere başlığı ayarlanıyor.
screen.bgcolor("black")  # Arka plan rengi siyah yapılıyor.
screen.setup(width=screen_width, height=screen_height)  # Pencere boyutları ayarlanıyor.
screen.tracer(
    0
)  # Otomatik ekran güncellemeleri kapatılıyor. Bu, animasyonun daha akıcı olmasını sağlar.

# --- Yılanın Kafasını Oluşturma ---
head = turtle.Turtle()  # Yılanın kafası için bir turtle nesnesi oluşturuluyor.
head.speed(0)  # Çizim hızını en yükseğe ayarlıyoruz.
head.shape("circle")  # Kafa şeklini daire yapıyoruz.
head.color("white")  # Rengini beyaz yapıyoruz.
head.penup()  # Hareket ederken arkasında iz bırakmasını engelliyoruz.
head.goto(0, 0)  # Başlangıç pozisyonu ekranın ortası.
head.direction = "stop"  # Oyun başında hareketsiz durması için yönünü "stop" yapıyoruz.

# --- Yemi Oluşturma ---
food = turtle.Turtle()  # Yem için bir turtle nesnesi oluşturuluyor.
food.speed(0)
food.shape("circle")
food.color("red")
food.penup()

# Yemin başlangıç pozisyonunu ekran sınırları içinde rastgele belirleme
try:
    max_x = int(screen_width / 2 - segment_size * 2)
    max_y = int(screen_height / 2 - segment_size * 2)
    # Yılanın hareket adımlarına (segment_size) tam oturması için pozisyonu yuvarlıyoruz.
    x_init = round(random.randint(-max_x, max_x) / segment_size) * segment_size
    y_init = round(random.randint(-max_y, max_y) / segment_size) * segment_size
    food.goto(x_init, y_init)
except ValueError:
    food.goto(0, 100)  # Olası bir hatada varsayılan konuma gönder.

# Yılanın kuyruk segmentlerini saklamak için bir liste oluşturuyoruz.
segments = []

# --- Skor Tabelasını Oluşturma ---
score_display = turtle.Turtle()
score_display.speed(0)
score_display.shape("square")  # Şeklinin bir önemi yok çünkü gizlenecek.
score_display.color("white")
score_display.penup()
score_display.hideturtle()  # Turtle ikonunu (imleci) gizliyoruz.
score_display.goto(0, screen_height / 2 - 40)  # Ekranın üst kısmına konumlandırıyoruz.
score_display.write(
    "Skor: 0  Yüksek Skor: 0", align="center", font=("Courier", 24, "normal")
)


# --- Yön Kontrol Fonksiyonları ---
# Bu fonksiyonlar, yılanın mevcut yönünün tersine gitmesini engeller.
def go_up():
    if head.direction != "down":
        head.direction = "up"


def go_down():
    if head.direction != "up":
        head.direction = "down"


def go_left():
    if head.direction != "right":
        head.direction = "left"


def go_right():
    if head.direction != "left":
        head.direction = "right"


# --- Yılanı Hareket Ettirme Fonksiyonu ---
def move():
    # Kayanın yönü ne ise, o yöne doğru 'segment_size' kadar ilerletir.
    if head.direction == "up":
        head.sety(head.ycor() + segment_size)
    if head.direction == "down":
        head.sety(head.ycor() - segment_size)
    if head.direction == "left":
        head.setx(head.xcor() - segment_size)
    if head.direction == "right":
        head.setx(head.xcor() + segment_size)


# --- Klavye Tuşlarını Dinleme ---
screen.listen()  # Klavyeden gelen girdileri dinlemeye başla.
# Belirtilen tuşlara basıldığında ilgili yön fonksiyonunu çağır.
screen.onkeypress(go_up, "w")
screen.onkeypress(go_down, "s")
screen.onkeypress(go_left, "a")
screen.onkeypress(go_right, "d")
# Ok tuşları için de aynı atamaları yapıyoruz.
screen.onkeypress(go_up, "Up")
screen.onkeypress(go_down, "Down")
screen.onkeypress(go_left, "Left")
screen.onkeypress(go_right, "Right")


# --- Oyunu Sıfırlama Fonksiyonu ---
# Oyuncu yandığında çağrılır, kod tekrarını önler.
def reset_game():
    global score, delay
    time.sleep(1)  # Oyuncu ne olduğunu anlasın diye 1 saniye bekle.
    head.goto(0, 0)
    head.direction = "stop"

    # Tüm segmentleri ekranın dışına taşıyarak temizle.
    for segment in segments:
        segment.goto(1000, 1000)
    segments.clear()  # Segment listesini boşalt.

    score = 0  # Skoru sıfırla.
    delay = 0.1  # Hızı başlangıç değerine döndür.

    # Skor tabelasını güncelle.
    score_display.clear()
    score_display.write(
        "Skor: {}  Yüksek Skor: {}".format(score, high_score),
        align="center",
        font=("Courier", 24, "normal"),
    )


# +++++++++++++++++ ANA OYUN DÖNGÜSÜ +++++++++++++++++
while True:
    screen.update()  # Ekranı her döngüde manuel olarak güncelliyoruz.

    # Kenarlara çarpma kontrolü
    if (
        head.xcor() > screen_width / 2 - 10
        or head.xcor() < -screen_width / 2 + 10
        or head.ycor() > screen_height / 2 - 10
        or head.ycor() < -screen_height / 2 + 10
    ):
        reset_game()  # Yılan kenara çarparsa oyunu sıfırla.

    # Yem yeme kontrolü
    if (
        head.distance(food) < 20
    ):  # Yılanın kafası ile yem arasındaki mesafe 20 pikselden azsa...
        # Yem için yeni bir rastgele konum belirle.
        x = (
            round(
                random.randint(
                    int(-screen_width / 2 + segment_size),
                    int(screen_width / 2 - segment_size),
                )
                / segment_size
            )
            * segment_size
        )
        y = (
            round(
                random.randint(
                    int(-screen_height / 2 + segment_size),
                    int(screen_height / 2 - segment_size),
                )
                / segment_size
            )
            * segment_size
        )
        food.goto(x, y)

        # Yılanın kuyruğuna yeni bir segment ekle.
        new_segment = turtle.Turtle()
        new_segment.speed(0)
        new_segment.shape("circle")
        new_segment.color("grey")
        new_segment.penup()
        segments.append(new_segment)

        delay -= 0.001  # Her yem yediğinde oyunu bir miktar hızlandır.

        # Skoru arttır ve tabelayı güncelle.
        score += 10
        if score > high_score:
            high_score = score
        score_display.clear()
        score_display.write(
            "Skor: {}  Yüksek Skor: {}".format(score, high_score),
            align="center",
            font=("Courier", 24, "normal"),
        )

    # Segmentleri hareket ettirme
    # Her segmenti, bir önceki segmentin pozisyonuna taşıyoruz.
    # Döngü sondan başa doğru çalışır.
    for index in range(len(segments) - 1, 0, -1):
        x = segments[index - 1].xcor()
        y = segments[index - 1].ycor()
        segments[index].goto(x, y)

    # Kuyruğun ilk parçasını yılanın kafasının eski pozisyonuna taşı.
    if len(segments) > 0:
        x = head.xcor()
        y = head.ycor()
        segments[0].goto(x, y)

    move()  # Yılanın kafasını yeni pozisyonuna hareket ettir.

    # Kuyruğa çarpma kontrolü
    for segment in segments:
        if segment.distance(head) < 20:  # Kafa, segmentlerden birine çok yaklaşırsa...
            reset_game()  # Oyunu sıfırla.

    time.sleep(
        delay
    )  # Döngüyü 'delay' süresi kadar beklet. Bu, oyunun hızını belirler.

# Bu komut, program bitse bile pencerenin kapanmamasını sağlar.
# Sonsuz döngü olduğu için bu koda ulaşılamaz ama turtle programlarında kullanmak iyi bir alışkanlıktır.
turtle.done()
```
